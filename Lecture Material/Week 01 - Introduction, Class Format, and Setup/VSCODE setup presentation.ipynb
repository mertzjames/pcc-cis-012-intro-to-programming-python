`CTRL+SHIFT+P`
- type 'install'
- select 'Extensions: Install Extensions'
- search for 'Python'
    - be sure to check that the publisher is Microsoft
- install the python extension
    - this may require you to restart VS Code


- type and select 'python' as your file type

`CTRL+SHIFT+P` again
- type 'interpreter'
- select 'Python: Select Interpreter'
- select the interpreter on your machine
- make sure that you see your interpreter in the bottom left of your VScode Window


- type the standard 'hello-world' code and save your file
